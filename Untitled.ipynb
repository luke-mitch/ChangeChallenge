```python
from itertools import count
```

```python
a=1
b=2
c=5
d=10
e=20
f=50
g=100

TOTAL = 200
```

```python
PERM = 0

for h in range(0, 1+200//a):
    remainder_a = TOTAL - a*h
    for i in range(0, 1+remainder_a//b):
        remainder_b = remainder_a - b*i
        for j in range(0, 1+remainder_b//c):
            remainder_c = remainder_b - c*j
            for k in range(0, 1+remainder_c//d):
                remainder_d = remainder_c - d*k
                for l in range(0, 1+remainder_d//e):
                    remainder_e = remainder_d - e*l
                    for m in range(0, 1+remainder_e//f):
                        remainder_f = remainder_e - f*m
                        for n in range(0, 1+remainder_f//g):
                            remainder_g = remainder_f - g*n
                            if a*h+b*i+c*j+d*k+e*l+f*m+g*n == 200: 
                                print (h, i, j, k, l, m, n)
                                PERM = PERM+1
                                break
                            elif a*h+b*i+c*j+d*k+e*l+f*m+g*n > 200: 
                                break
                                
print (PERM)
```

```python
results=[]

for h in count(0):
    for i in count(0):
        for j in count(0):
            for k in count(0):
                for l in count(0):
                    for m in count(0):
                        for n in count(0):
                            if n * 1 + m * 2 + l * 5 + k * 10 + j * 20 + i * 50 + h * 100 == TOTAL:
                                results.append({100: h, 50: i, 20: j, 10: k, 5: l, 2: m, 1: n})
                            if n * 1 + m * 2 + l * 5 + k * 10 + j * 20 + i * 50 + h * 100 > TOTAL:
                                break
                            if n * 1 + m * 2 + l * 5 + k * 10 + j * 20 + i * 50 > TOTAL:
                                break
                            if n * 1 + m * 2 + l * 5 + k * 10 + j * 20 > TOTAL:
                                break
                            if n * 1 + m * 2 + l * 5 + k * 10 > TOTAL:
                                break
                            if n * 1 + m * 2 + l * 5 > TOTAL:
                                break    
                            if n * 1 + m * 2 > TOTAL:
                                break
                            if n * 1 > TOTAL:
                                break
                            

print (results)
```
